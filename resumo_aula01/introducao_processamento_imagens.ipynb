## Diferença entre síntese de imagem, processamento de imagem e visão computacional

A diferença entre sintese de imagens, processamento de imagens e visão computacional, está na entrada e saída dos dados. Na síntese de imagens, tem-se como dados de entradas modelos e dados e na sua saída uma imagem, no processamento digital de imagens, tem-se como dado de entrada uma imagem e na sua saída uma outra imagem. Por último, na visão computacional, a entrada é uma imagem e a sua saída são modelos e dados.

## Sistema visual humano

### As 3 principais membranas do olho humano

As membranas dos olhos humanos que atual principalmente como camadas de proteção são: 
- Esclerótica, camada mas externa
- Coróide, membrana intermediária
- Retina, membrana interna

### Córnea

A córnea atua como uma lente refratora da luz no olho humano 

### Irís

A íris é responsável por controlar a entrada de luz no olho humano, em ambientes escuros ela dilata enquanto em ambientes muito claro ela tende a fechar a abertura para que pouca
luz entre e não prejudique nossos olhos.

### Cristalino

O cristalino atua junto com a córnea e ajudam a focalizar a luz para a formação da imagem no fundo dos olhos.

### Retina

A retina é responsável por codificar as informações e enviar para o cérebro humano. Ela é composta por células sensíveis a luz: os cones e os bastonetes. 


- Os cones são muito sensíveis a luz e possuem 3 tipos, cada um sensível ao comprimento de onda de uma das 3 cores RGB.


- Os bastonetes são células sensíveis a baixas intensidades luminosas.

## Modelo de Imagem

Uma imagem é definidacomo: f(x,y);
    
- Cada posição (x,y) corresponde a uma intensidade ou brilho;


- Quantidade de luz incidente (iluminância);


- Quantidade de luz refletida pelos objetos na cena(Reflectância);

Alguns valores de lúmen por metro quadrado comuns.

## Digitalização de Imagens

No processo de aquisição de imagens, uma fonte de luz incide raios luminosos no elemento em cena, esses raios são refletidos e um sistema de aquisição de imagens projeta os raios refratados em um plano interno de aquisição de imagens, a é então imagem digitalizada.

### Amostragem

Consiste em discretizar o domínio de definição da imagem nas direções x e y, gerando uma matriz de M x N amostras;


Basicamente, a amostragem converte a imagem analógica em uma matriz de M x N pontos, denominados pixels.

### Quantização

Etapa responsável por fazer com que cada um destes pixels assuma uma valor inteiro;


0 – 2 <sup>n</sup> – 1;


Quanto maior o valor de n, maior será o número de níveis de cinza presentes na imagem digitalizada.
